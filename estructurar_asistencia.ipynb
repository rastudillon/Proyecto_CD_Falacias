```python
#2-1

import os

def clean_file(filename):
  """
  This function cleans a single file by removing lines starting with "SESIÓN" 
  and the following line, as well as lines containing "CÁMARA DE DIPUTADOS" 
  and the following line.

  Args:
      filename: The name of the file to be cleaned.
  """
  with open(filename, 'r', encoding="utf-8") as f:
    lines = f.readlines()

  cleaned_lines = []
  skip_next = False
  for line in lines:
    if line.startswith("SESIÓN"):
      skip_next = True
    elif "CÁMARA DE DIPUTADOS" in line:
      skip_next = True
    elif not skip_next:
      cleaned_lines.append(line)
    else:
      skip_next = False

  with open(filename, 'w') as f:
    f.writelines(cleaned_lines)


def clean_folder(folder_path):
  """
  This function iterates through all files in a folder and calls clean_file 
  on each text file.

  Args:
      folder_path: The path to the folder containing the text files.
  """
  for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
      clean_file(os.path.join(folder_path, filename))


# Replace "path/to/your/folder" with the actual path to your folder
clean_folder(r'C:\Users\USUARIO\Mi unidad (ramon.astudillo.nahuelcoy@alumnos.uta.cl)\Proyecto Ciencia de Datos II\PDF Sesiones camara de Diputados\Legislatura 335\asistencia\asist_txt')

print("Files cleaned successfully!")
```

```python
#2 leg 336

import os
from natsort import natsorted

def clean_file(filename):
    """
    This function cleans a single file by removing lines starting with "SESIÓN"
    and the following line, as well as lines containing "CÁMARA DE DIPUTADOS"
    and the following line. Additionally, it removes lines containing specific keywords and replaces commas with spaces.
    
    Args:
        filename: The name of the file to be cleaned.
    """
    with open(filename, 'r', encoding="utf-8") as f:
        lines = f.readlines()

    cleaned_lines = []
    skip_next = False
    for line in lines:
        if line.startswith("SESIÓN") or line.startswith("CÁMARA"):
            skip_next = True
        elif skip_next:
            skip_next = False
        elif ("NOMBRE" in line or "(Partido" in line or "Región" in line or 
              "Distrito" in line or "*" in line):
            continue  # Saltar esta línea
        else:
            # Reemplazar comas con espacios
            cleaned_line = line.replace(",", " ").strip()
            cleaned_lines.append(cleaned_line)

    # Eliminar líneas vacías y unir las líneas limpias en un solo string
    cleaned_content = '\n'.join(cleaned_lines).strip()

    with open(filename, 'w', encoding="utf-8") as f:
        f.write(cleaned_content + '\n')  # Agregar una nueva línea al final

def clean_folder(folder_path):
    """
    This function iterates through all files in a folder and calls clean_file
    on each text file.
    
    Args:
        folder_path: The path to the folder containing the text files.
    """
    # Listar todos los archivos .txt en el directorio
    txt_files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]

    # Ordenar los archivos utilizando natsort para una ordenación natural
    txt_files = natsorted(txt_files)

    # Limpiar cada archivo en el directorio
    for txt_file in txt_files:
        file_path = os.path.join(folder_path, txt_file)
        clean_file(file_path)
        print(f"El archivo '{txt_file}' ha sido limpiado exitosamente.")

# Reemplaza "path/to/your/folder" con la ruta actual a tu carpeta
clean_folder(r'C:\Users\USUARIO\Mi unidad (ramon.astudillo.nahuelcoy@alumnos.uta.cl)\Proyecto Ciencia de Datos II\PDF Sesiones camara de Diputados\Legislatura 336\asistencia\asist_txt')

print("Archivos limpiados exitosamente!")
```
