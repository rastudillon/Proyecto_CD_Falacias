```python
import fitz  #pip install PyMuPDF
import os

def convert_pdf_to_txt_pyMuPDF(pdf_path, txt_path):
    doc = fitz.open(pdf_path)
    with open(txt_path, 'w', encoding='utf-8') as txt_file:
        for page in doc:
            txt_file.write(page.get_text())

def convert_all_pdfs_in_directory(base_directory, start_leg, end_leg):
    # Asumiendo que los directorios están nombrados como "Legislatura XXX"
    for i in range(start_leg, end_leg + 1):
        current_dir = os.path.join(base_directory, f"Legislatura {i}")
        if os.path.exists(current_dir):
            for filename in os.listdir(current_dir):
                if filename.endswith('.pdf'):
                    pdf_path = os.path.join(current_dir, filename)
                    txt_path = os.path.join(current_dir, os.path.splitext(filename)[0] + '.txt')
                    convert_pdf_to_txt_pyMuPDF(pdf_path, txt_path)
                    print(f"Convertido: {filename} a TXT en {current_dir}")
        else:
            print(f"Directorio no encontrado: {current_dir}")

# Ruta al directorio base que contiene los subdirectorios de cada legislatura
base_directory = r"C:\Users\USUARIO\Mi unidad (ramon.astudillo.nahuelcoy@alumnos.uta.cl)\Proyecto Ciencia de Datos II\PDF Sesiones camara de Diputados"

# Llama a la función con el rango de legislaturas deseado
convert_all_pdfs_in_directory(base_directory, 319, 372)
```
